In this [paper](https://www.microsoft.com/en-us/research/wp-content/uploads/2016/02/Ferreira_Fisher_Sample_Oriented_Tasks.pdf) the authors describe the challenges users face when trying to make judgements about probabilistic data generated through samples. As an example, they look at a bar chart of four years of data (replicated below in Figure 1). Each year has a y-axis value, which is derived from a sample of a larger dataset. For instance, the first value might be the number votes in a given district or riding for 1992, with the average being around 33,000. On top of this is plotted the 95% confidence interval for the mean (see the boxplot lectures for more information, and the yerr parameter of barcharts).

A challenge that users face is that, for a given y-axis value (e.g. 42,000), it is difficult to know which x-axis values are most likely to be representative, because the confidence levels overlap and their distributions are different (the lengths of the confidence interval bars are unequal). One of the solutions the authors propose for this problem (Figure 2c) is to allow users to indicate the y-axis value of interest (e.g. 42,000) and then draw a horizontal line and color bars based on this value. So bars might be colored red if they are definitely above this value (given the confidence interval), blue if they are definitely below this value, or white if they contain this value.

```python
# Use the following data for this assignment:

import pandas as pd
import numpy as np

np.random.seed(12345)

df = pd.DataFrame([np.random.normal(32000,200000,3650), 
                   np.random.normal(43000,100000,3650), 
                   np.random.normal(43500,140000,3650), 
                   np.random.normal(48000,70000,3650)], 
                  index=[1992,1993,1994,1995])
df
```

## First create the bar chart for my dataframe. This visualizes the mean and 95% confidence interval for each of the data samples (n = 3650) of the normal distributions

```python
%matplotlib notebook
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# first convert these four data samples of normal distributions into a bar chart. 
df_copy = df.transpose()

# get values of means, population std, sample std, etc...
value = df_copy.mean()
std = df_copy.std()
sample_std = (std / np.sqrt(3650)) 
z_value = (stats.t.ppf(1-0.05/2, 3650 - 1))
half_confidence_interval_95 = (sample_std) * (stats.t.ppf(1-0.05/2, 3650 - 1))

# Create a new axes on a figure
#fig1, ax1 = plt.subplots()
# Basic  bar chart based on datasample parameters above
#plt.bar(left = range(len(df_copy.columns)), height = value, bottom = 0, align='center',
#    yerr = half_confidence_interval_95, capsize = 12, tick_label = df_copy.columns)
```

### Easiest option: Implement the bar coloring as described above - a color scale with only three colors, (e.g. blue, white, and red). Assume the user provides the y axis value of interest as a parameter or variable.

```python
# simple function for user to add horizontal line onto the bar chart
def add_horizontal_line(y_value):
    
    # Create a new axes on a figure
    fig2, ax2 = plt.subplots(figsize=(8,5))
    
    # plot the horizontal line
    plt.axhline(y = y_value, color = 'grey')
    
    # Create a list of confidence intervals by order of dataframe index
    confidence_intervals_upper_boundary = value + half_confidence_interval_95
    confidence_intervals_lower_boundary = value - half_confidence_interval_95
    confidence_intervals = list(zip(confidence_intervals_lower_boundary, confidence_intervals_upper_boundary))
    
    # list of colors for the bars. This is based on the conditions of comparing 
    # the hline value to the confidence interval boundary values
    colors = ['red' if (y_value < x[0]) else ('blue' if (y_value > x[1]) else 'white') for x in confidence_intervals]
    
    # Plot the bar chart of mean and confidence intervals of the dataframe, now with color coding depending on the 
    # relationship between hline and confidence interval boundary values
    
    plt.bar(left = range(len(df_copy.columns)), height = value, bottom = 0, align='center',
        yerr = half_confidence_interval_95, color = colors, edgecolor = 'black', capsize = 12, tick_label = df_copy.columns)
    
    # add axes title and axis labels
    plt.xlabel('Year')
    plt.ylabel('Sample Data Mean')
    plt.title('Means of datasamples for each year \n with 95% confidence intervals', fontsize = 10)
```

```python
# User can provide a value to visualize hline, bar chart, and color coding of the bar chart
# using add_horizontal_line function

# Example:
add_horizontal_line(41000)
```

```python
# add legend
colors = {'hline beneath confidence interval':'red', 'hline above confidence interval':'blue',
          'hline between confidence interval': 'white'}         
labels = list(colors.keys())
handles = [plt.Rectangle((0,0),1,1, color=colors[label]) for label in labels]
plt.legend(handles, labels, bbox_to_anchor=(1, 1), loc='upper left', prop={'size': 7})
# adjust the subplot so the legend doesn't run off the image
plt.subplots_adjust(top = 0.85, bottom=0.15, left = 0.11, right = 0.65)
```
